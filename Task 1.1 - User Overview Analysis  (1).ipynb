# Task 1.1

#### Aggregate per user the following information in the columnÂ¶
Number of xDR sessions

```python
sessions = telcom_data.groupby('MSISDN/Number').agg({'Bearer Id': 'count'})
sessions = sessions.rename(columns= {'Bearer Id': 'xDR_sessions'})
sessions.sort_values(by=['xDR_sessions'], ascending = False).head(10)
```

```python
sessions.describe()
```

#### Session duration

```python
# we group the data using  msisdn (customer number) and find sum the the duration for each customer

durations = telcom_data.groupby('MSISDN/Number').agg({'Dur. (ms)': 'sum'})


durations=durations.rename(columns= {'Dur. (ms)': 'total_duration(ms)'})
durations.sort_values(by=['total_duration(ms)'], ascending = False).head(10)
```

```python
durations.describe()
```

#### the total download (DL) and upload (UL) data

```python
total_data_df = telcom_data[['MSISDN/Number','Total UL (Bytes)','Total DL (Bytes)']].copy()
total_data_df = total_data_df.groupby('MSISDN/Number').sum()
total_data_df['total_data(bytes)'] = total_data_df['Total UL (Bytes)'] + total_data_df['Total DL (Bytes)']
total_data_df.sort_values(by=['total_data(bytes)'], ascending = False).head(10)
```

```python
total_data_df.describe()
```

#### the total data volume (in Bytes) during this session for each application

first lets commute the total data (ul + dl) for each application. and added new columns to our data (Social Media, Google, Email, Youtube, Netflix, Gaming, and Other.


```python
telcom_data["Social_Media"] = telcom_data["Social Media DL (Bytes)"] + telcom_data['Social Media UL (Bytes)']
telcom_data["Google"] = telcom_data["Google DL (Bytes)"] + telcom_data["Google UL (Bytes)"]
telcom_data['Email'] = telcom_data["Email DL (Bytes)"] + telcom_data["Email UL (Bytes)"]
telcom_data['Youtube'] = telcom_data["Youtube DL (Bytes)"] + telcom_data["Youtube UL (Bytes)"]
telcom_data['Netflix'] = telcom_data["Netflix DL (Bytes)"] + telcom_data["Netflix UL (Bytes)"]
telcom_data["Gaming"] = telcom_data["Gaming DL (Bytes)"] + telcom_data["Gaming UL (Bytes)"]
telcom_data['Other']= telcom_data["Other DL (Bytes)"]+ telcom_data["Other UL (Bytes)"]
telcom_data['Total_data'] = telcom_data['Total DL (Bytes)'] + telcom_data['Total UL (Bytes)']
```

```python
app_total_data = telcom_data[['MSISDN/Number','Social_Media','Google',
                             'Email', 'Youtube','Netflix', 'Gaming','Other']].copy()
```
